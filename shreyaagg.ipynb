```python
from sklearn.preprocessing import StandardScaler
```

```python
print(df.isnull().sum())
```

```python
df = df.drop_duplicates()
```

```python
df['diabetes'].value_counts()
```

```python
df.groupby('diabetes').mean(numeric_only=True)
```

```python
columns_to_drop = ['diabetes', 'gender', 'smoking_history']
existing_columns_to_drop = [col for col in columns_to_drop if col in df.columns]
x = df.drop(columns=existing_columns_to_drop, axis=1)
```

```python
print(x)
```

```python
y = df['diabetes']
```

```python
print(y)
```

```python
scalar = StandardScaler()
```

```python
scalar.fit(x)
```

```python
s_d = scalar.transform(x)
```

```python
print(s_d)
```

```python
x = s_d
```

```python
y = df['diabetes']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, stratify = y, random_state = 2)
```

model training

```python
from sklearn import svm
```

```python
classifier = svm.SVC(kernel = 'linear')
```

```python
classifier.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
x_train_prediction = classifier.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print("accuracy score of training dataset : ", training_data_accuracy, y_train)
```
